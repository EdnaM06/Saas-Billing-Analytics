## Overview

This project analyzes subscription billing and revenue data to simulate the work of a Billing Data Analyst in a modern SaaS environment. The focus is on understanding how customer behavior, billing performance, and product interactions impact overall financial health.

The analysis explores how data can be used to improve:

- Revenue performance and monthly recurring revenue (MRR)
- Customer retention and churn reduction
- Billing reliability and payment success rates
- Risk detection in subscription and payment behavior
- Experimentation outcomes through A/B testing

The objective is to replicate how billing and revenue data is used to support **financial decision-making, operational efficiency, and risk mitigation** in a subscription-based business.

## Problem Statement

Subscription-based businesses rely on recurring revenue, which makes financial stability highly sensitive to customer behavior and billing system performance. However, several operational challenges can directly impact revenue consistency and growth.

A key issue is **billing inefficiency**, including failed payments, delayed transactions, and system-level errors that result in **revenue leakage and lost income opportunities**. Alongside this, **customer churn_both voluntary cancellations and involuntary losses due to payment failures—reduces recurring revenue and weakens long-term customer value.**

Another challenge is the limited visibility into how product changes, pricing decisions, and marketing campaigns influence revenue outcomes. Without clear analytical feedback loops, it becomes difficult to determine whether growth is driven by improved conversion or offset by higher churn and customer dissatisfaction.

Finally, many subscription systems lack strong **predictive and scenario-based analytics**, making it harder to anticipate the financial impact of changes in churn rates, pricing strategies, or billing performance.

This analysis addresses these challenges by focusing on **revenue performance, churn behavior, billing efficiency, and risk patterns**, with the goal of generating insights that support better financial decision-making and operational improvement.

## Simulated A/B Test

```python
# 1. Create A/B test groups

np.random.seed(42)

df_subscriptions['group'] = np.random.choice(['A', 'B'], size=len(df_subscriptions))

print(df_subscriptions['group'].value_counts())
```

```python
# 2. Define conversion column

df_subscriptions['converted'] = (df_subscriptions['subscription_status'] == 'paid').astype(int)

print(df_subscriptions['converted'].value_counts())
```

```python
# 3. Split A and B groups

group_a = df_subscriptions[df_subscriptions['group'] == 'A']
group_b = df_subscriptions[df_subscriptions['group'] == 'B']
```

```python
# 4. Conversion rates

conversion_a = group_a['converted'].mean()
conversion_b = group_b['converted'].mean()

print("Group A Conversion:", conversion_a)
print("Group B Conversion:", conversion_b)
```

## Statistical significance

```python
# 5. Statistical test (t-test)

t_stat, p_value = stats.ttest_ind(group_a['converted'], group_b['converted'])

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
# 6. Interpretation of results

if p_value < 0.05:
    print("Result is statistically significant → difference is real")
else:
    print("No statistical significance → difference may be random")
```
