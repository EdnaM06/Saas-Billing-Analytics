## Overview

This project analyzes subscription billing and revenue data to simulate the work of a Billing Data Analyst in a modern SaaS environment. The focus is on understanding how customer behavior, billing performance, and product interactions impact overall financial health.

The analysis explores how data can be used to improve:

- Revenue performance and monthly recurring revenue (MRR)
- Customer retention and churn reduction
- Billing reliability and payment success rates
- Risk detection in subscription and payment behavior
- Experimentation outcomes through A/B testing

The objective is to replicate how billing and revenue data is used to support **financial decision-making, operational efficiency, and risk mitigation** in a subscription-based business.

## Problem Statement

Subscription-based businesses rely on recurring revenue, which makes financial stability highly sensitive to customer behavior and billing system performance. However, several operational challenges can directly impact revenue consistency and growth.

A key issue is **billing inefficiency**, including failed payments, delayed transactions, and system-level errors that result in **revenue leakage and lost income opportunities**. Alongside this, **customer churn_both voluntary cancellations and involuntary losses due to payment failures—reduces recurring revenue and weakens long-term customer value.**

Another challenge is the limited visibility into how product changes, pricing decisions, and marketing campaigns influence revenue outcomes. Without clear analytical feedback loops, it becomes difficult to determine whether growth is driven by improved conversion or offset by higher churn and customer dissatisfaction.

Finally, many subscription systems lack strong **predictive and scenario-based analytics**, making it harder to anticipate the financial impact of changes in churn rates, pricing strategies, or billing performance.

This analysis addresses these challenges by focusing on **revenue performance, churn behavior, billing efficiency, and risk patterns**, with the goal of generating insights that support better financial decision-making and operational improvement.

## A/B Testing

```python
# Create A/B test groups

np.random.seed(42)

df_subscriptions['group'] = np.random.choice(['A', 'B'], size=len(df_subscriptions))

print(df_subscriptions['group'].value_counts())
```

```python
# Define conversion column

df_subscriptions['group'] = np.random.choice(['A', 'B'], size=len(df_subscriptions))
```

```python
# Split A and B groups

group_a = df_subscriptions[df_subscriptions['group'] == 'A']
group_b = df_subscriptions[df_subscriptions['group'] == 'B']
```

```python
# Conversion rates

group_a = df_subscriptions[df_subscriptions['group'] == 'A']
group_b = df_subscriptions[df_subscriptions['group'] == 'B']

conversion_a = group_a['converted'].mean()
conversion_b = group_b['converted'].mean()

print("A:", conversion_a)
print("B:", conversion_b)
```

```python
# Hypothesis test

t_stat, p_value = stats.ttest_ind(
    group_a['converted'],
    group_b['converted'],
    equal_var=False
)

print("T-stat:", t_stat)
print("P-value:", p_value)
```

### Business Recommendation:

The A/B test shows no statistically significant difference in conversion rates between Group A and Group B (p = 0.21 > 0.05).

Although Group B shows a slightly higher conversion rate, the difference is not strong enough to conclude that it improves performance.

Recommendation:
- Do not roll out Group B based on current results.
- Maintain the existing version (Group A) as the default experience.
- Consider running a longer test or increasing sample size to improve statistical power before making a final decision.

## Statistical significance

```python
# Statistical test (t-test)

t_stat, p_value = stats.ttest_ind(group_a['converted'], group_b['converted'])

print("T-stat:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Significant difference")
```

```python
# Interpretation of results

if p_value < 0.05:
    print("Result is statistically significant → difference is real")
else:
    print("No statistical significance → difference may be random")
```

## Key Business Insights

**1. Revenue Stability Depends on Retention**

Revenue trends show moderate fluctuations over time, but no strong sustained growth pattern. This indicates that retention has a stronger influence on revenue stability than new customer acquisition.

- Business implication: Improving retention will yield higher ROI than focusing only on acquisition.

**2. Churn is the Primary Business Risk**

The churn rate (~26%) is relatively high for a SaaS model, meaning a significant portion of users are leaving the platform within a short period.

- Business implication: Customers are not retaining long enough to maximize lifetime value, leading to revenue leakage.

**3. A/B Test Results Are Not Statistically Significant**

The difference in conversion rates between Group A and B is not statistically significant (p-value > 0.05), meaning the observed difference is likely due to random variation.

- Business implication: No product change should be rolled out based on this experiment.

**4. Revenue Is Unevenly Distributed Across Plans**

Revenue is concentrated in specific subscription tiers, indicating that higher-value plans contribute disproportionately to total revenue.

- Business implication: Upselling strategies may be more effective than increasing low-tier acquisition.

**5. Billing Risk Exists but Is Not the Main Driver**

A small subset of users classified as high-risk (free users who churn) indicates potential billing inefficiencies, but this is not the primary revenue driver issue.

- Business implication: Billing optimization is secondary to churn reduction.

## Conclusion 

This analysis of SaaS subscription data revealed that customer retention is the most critical driver of revenue stability, with churn rates (~26%) posing the highest business risk.

While conversion rates remain relatively stable across user segments, the lack of statistically significant difference in A/B testing results indicates that current product variations do not materially impact user behavior.

Revenue performance is primarily influenced by subscription plan distribution and user retention patterns rather than short-term acquisition changes.

**Key takeaway:**

- Improving retention and reducing churn will have a significantly higher impact on revenue growth than optimizing conversion alone.

**Next steps:**

- Focus on churn reduction strategies (onboarding, engagement, pricing review)
- Improve A/B testing design with larger samples and clearer hypotheses
- Build predictive churn models for early intervention
